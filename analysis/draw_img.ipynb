```python
import matplotlib.pyplot as plt
import matplotlib as mpl

# 全局美化设置
mpl.rcParams['font.size'] = 12
mpl.rcParams['axes.titlesize'] = 18

# 数据
labels = ['lumbar_pelvis', 'head', 'shoulder', 'foot', 'wrist']
sizes = [78, 43, 38, 1, 1]
colors = ['green', 'orange', 'red', 'purple', 'blue']  # 保留自定义颜色

# 绘图
fig, ax = plt.subplots(figsize=(6, 6))
wedges, texts, autotexts = ax.pie(
    sizes,
    labels=labels,
    autopct=lambda p: f'{p:.1f}%\n({int(p * sum(sizes) / 100)})',
    colors=colors,
    startangle=90,
    wedgeprops=dict(edgecolor='white', linewidth=2),
    textprops=dict(color='black')
)

# 百分比数字美化
for autotext in autotexts:
    autotext.set_color('white')  # 白色字体在有色区域上更显眼
    autotext.set_fontsize(11)
    autotext.set_weight('bold')

# 标题
ax.set_title('Doctor 1', weight='bold')
plt.axis('equal')  # 保持圆形
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl

# 美化设置
mpl.rcParams['font.size'] = 12
mpl.rcParams['axes.titlesize'] = 18

# 数据
labels = ['lumbar_pelvis', 'head', 'shoulder', 'wrist', 'foot']
sizes = [98, 57, 4, 3, 2]
colors = plt.cm.tab10.colors[:len(labels)]  # 使用更专业的 tab10 色板

# 画图
fig, ax = plt.subplots(figsize=(6, 6))
wedges, texts, autotexts = ax.pie(
    sizes,
    labels=labels,
    autopct=lambda p: f'{p:.1f}%\n({int(p * sum(sizes) / 100)})',
    colors=colors,
    startangle=90,
    wedgeprops=dict(edgecolor='white', linewidth=2),
    textprops=dict(color='black')
)

# 居中的白色百分比文字（更明显）
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontsize(11)
    autotext.set_weight('bold')

# 设置标题
ax.set_title('Doctor 2', weight='bold')
plt.axis('equal')  # 保持为圆
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# 数据
doctor1 = {'foot': 0, 'head': 43, 'lumbar_pelvis': 78, 'shoulder': 38, 'wrist': 3}
doctor2 = {'foot': 2, 'head': 57, 'lumbar_pelvis': 98, 'shoulder': 4, 'wrist': 1}

labels = list(doctor1.keys())
values1 = [doctor1[k] for k in labels]
values2 = [doctor2[k] for k in labels]

x = range(len(labels))
width = 0.35

# 创建图表
fig, ax = plt.subplots(figsize=(15, 5))

# 条形图
bars1 = ax.bar([i - width/2 for i in x], values1, width, label='Doctor 1', color='skyblue')
bars2 = ax.bar([i + width/2 for i in x], values2, width, label='Doctor 2', color='orange')

# 为每个柱子上加数字
ax.bar_label(bars1, padding=3, fontsize=10)
ax.bar_label(bars2, padding=3, fontsize=10)

# 轴设置
ax.set_ylabel('Count')
ax.set_title('Joint Selection by Doctor')
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=30)
ax.legend(loc='upper left')

# 表格数据
table_data = [[doctor1[k], doctor2[k]] for k in labels]

# 添加表格
table = plt.table(
    cellText=table_data,
    rowLabels=labels,
    colLabels=['Doctor 1', 'Doctor 2'],
    colWidths=[0.1]*2,
    loc='right',
    cellLoc='center',
    bbox=[1.05, 0.1, 0.3, 0.8]
)

table.auto_set_font_size(False)
table.set_fontsize(10)

plt.tight_layout()
plt.show()
```
